```python
data["Offer"] = data["Offer"].fillna(data["Offer"].mode()[0])
data["Multiple Lines"] = data["Multiple Lines"].fillna(data["Multiple Lines"].mode()[0])
data["Internet Type"] = data["Internet Type"].fillna(data["Internet Type"].mode()[0])
data["Online Security"] = data["Online Security"].fillna(data["Online Security"].mode()[0])
data["Online Backup"] = data["Online Backup"].fillna(data["Online Backup"].mode()[0])
data["Device Protection Plan"] = data["Device Protection Plan"].fillna(data["Device Protection Plan"].mode()[0])
data["Premium Tech Support"] = data["Premium Tech Support"].fillna(data["Premium Tech Support"].mode()[0])
data["Streaming TV"] = data["Streaming TV"].fillna(data["Streaming TV"].mode()[0])
data["Streaming Movies"] = data["Streaming Movies"].fillna(data["Streaming Movies"].mode()[0])
data["Streaming Music"] = data["Streaming Music"].fillna(data["Streaming Music"].mode()[0])
data["Unlimited Data"] = data["Unlimited Data"].fillna(data["Unlimited Data"].mode()[0])
data.fillna(data.select_dtypes(include="number").mean(), inplace=True)
```
